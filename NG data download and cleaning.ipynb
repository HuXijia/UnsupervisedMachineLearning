```python
def create_search(company, start, end, doctype):
    n = {"query": {
        "bool": {
            "must": [{
                
                "bool": {"should": 
                         [
                                {"match_phrase": {"text": company}},
                                {"match_phrase": {"title": company}},
                             ]}},
               
        {
          "match_phrase": {
            "doctype": doctype
          }},
            
        {"range" : {
            "publication_date" : {
                "gte" : start,
                "lte" : end,}}}
          
            
            ]}}}
    print(n)
    return n
```

```python
company = ['National Grid']
periods = [("2017-08-01", "2018-10-01")]
doctypes = ["dailymail", "guardian.com (print)", "guardian.com. (print)", "independent-uk (www)", "telegraph (www)", "the sun (print)", "thesun (www)","the daily telegraph (london) (print)","the daily telegraph(london) (print)","the guardian (london) (print)","the guardian (london) - final edition (print)","the guardian (print)","the guardian - final edition (print)","theguardian (www)","the sun (england) (print)"]  
```

```python
for company in company:
    for start, end in periods:
        for doctype in doctypes:
            print(company, start, end, doctype)
            n = create_search(company, start, end, doctype)
            myinca.importers_exporters.export_csv(query = n)
```
