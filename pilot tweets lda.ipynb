```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords 
from nltk.stem.wordnet import WordNetLemmatizer
import string
import re
```

```python
stop = set(stopwords.words('english'))
exclude = set(string.punctuation) 
lemma = WordNetLemmatizer()
def clean(text):
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', text)
    stop_free = " ".join([str(i) for i in cleantext.lower().split() if i not in stop])
    stop_free.replace('rt',' ')
#remove "nan" from text
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized

doc_clean = [clean(doc).split() for doc in text]
```

```python
doc_clean[10]
```

```python
import gensim
from gensim import corpora
```

```python
dictionary = corpora.Dictionary(doc_clean)
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
```

```python
Lda = gensim.models.ldamodel.LdaModel
```

```python
ldamodel = Lda(doc_term_matrix, num_topics=50, id2word = dictionary, passes=50)
```

```python
print(ldamodel.print_topics(num_topics=50, num_words=3))
```
